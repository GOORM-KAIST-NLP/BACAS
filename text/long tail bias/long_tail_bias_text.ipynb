```python
data = df[df['감정'] == '기쁨']
data
```

```python
# 기쁨 data 저장
#data.to_csv('/content/drive/MyDrive/Colab Notebooks/talk_data_happy_delete.xlsx')
```

```python
# back translation
from ktextaug import random_insert, random_delete, back_translate, TextAugmentation
from konlpy.tag import Okt, Mecab

agent = TextAugmentation(tokenizer="mecab", num_processes=1)
new_data = data.copy()

for i in range(len(data)):
    length = len(data)
    text = new_data.iloc[i]['원천데이터']

    a = agent.generate(text, mode="back_translate", target_language='ja')  
    b = back_translate(text_or_corpus=text, target_language='fr')
    c = back_translate(text_or_corpus=text, target_language='en')
    abc = pd.DataFrame({'감정' : ['기쁨', '기쁨', '기쁨'], '감정_소분류':['new_data', 'new_data', 'new_data'], '원천데이터':[a, b, c]})

    df = df.append(abc, ignore_index = True)
    df
```

```python
df = df.rename(columns = {'원천데이터':'증강데이터'})
df
```

```python
sns.countplot(x = '감정', data = df, order = df['감정'].value_counts().index)
plt.show()
```
